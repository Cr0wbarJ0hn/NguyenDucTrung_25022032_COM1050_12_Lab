```python
#W10A1
def divide(a,b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cant divide with number zero"
   

try:
    x,y = map(int,input().split())
    print(divide(x,y))
except ValueError:
    print("Please enter valid numbers")
    
```

```python
#W10A3
try:
    n = int(input())
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(d)**2 for d in str(n))
    
    if n == 1:
        print("YES")
    else:
        print("NO")
except ValueError:
    print("ValueError")
```

```python
def convert_list_items(lst):
    result = []
    for item in lst:
        try:
            val = int(item)
            result.append(val)
        except (ValueError, TypeError):
            print(f"Cảnh báo: Không thể chuyển đổi '{item}' sang số nguyên. Bỏ qua.")
    return result
```
