```python
#W6A1
def remove(arr):
    seen = []
    for number in arr:
        if number not in seen:
            seen.append(number)
    return seen
num_list = map(int,input().split())
print(list(remove(num_list)))
```

```python
#W6A2
def total_accumulated(arr):
    new = []
    k = 0
    list_arr = list(arr)
    while k < len(list_arr):
        new.append(sum(list_arr[:k + 1]))
        k +=1
    return new
num_list = map(int,input().split())
print(list(total_accumulated(num_list)))
```

```python
#W6A3
def move_left(tup, k):
    lst = list(tup)
    k = k % len(lst) if lst else 0  
    new_list = lst[k:] + lst[:k]
    return tuple(new_list)

num_list = tuple(map(int, input().split()))
x = int(input())
print(move_left(num_list, x))
```

```python
#W6A5
def classified(arr):
    d = {'positives': 0, 'negatives': 0, 'zeros': 0}
    for number in arr:
        if number > 0:
            d['positives'] += 1
        elif number < 0:
            d['negatives'] += 1
        else:
            d['zeros'] += 1
    return d

numb_list = list(map(int, input().split()))
print(classified(numb_list))
```

```python
#W6A9
def merge_dic(x, y):
    result = y.copy()
    for key in x:
        if key in result:
            result[key] = str(int(result[key]) + int(x[key]))
        else:
            result[key] = x[key]
    return dict(sorted(result.items()))


a = dict(pair.split(':') for pair in input().split())
b = dict(pair.split(':') for pair in input().split())
print(merge_dic(a, b))
```

```python
#W6A10
numbers = list(map(int, input().split()))
k = int(input())
pairs = []
seen = set()
for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] + numbers[j] == k:
            pair = (numbers[i], numbers[j])
            if pair not in seen:
                pairs.append(pair)
                seen.add(pair)
pairs.sort()
print(pairs)
```

```python
#W6A12
numbers = list(map(int, input().split()))
count = {}
for num in numbers:
    count[num] = count.get(num, 0) + 1
max_count = max(count.values())
result = min([num for num, freq in count.items() if freq == max_count])
print(result)
```

```python
def W6A16():
    n = int(input("Nhập số hàng (n): "))
    m = int(input("Nhập số cột (m): "))
    mang = []
    print("Nhập từng hàng, các phần tử cách nhau bởi dấu cách:")
    for i in range(n):
        hang = list(map(int, input(f"Hàng {i+1}: ").split()))
        mang.append(hang)
    print("\nMa trận đã nhập (định dạng 4 vị trí):")
    for hang in mang:
        print("".join(f"{phan_tu:4d}" for phan_tu in hang))
W6A16()
```

```python
def W6A17():
    n = int(input("Nhập kích thước ma trận vuông (n): "))
    mang = []
    print("Nhập từng hàng, các phần tử cách nhau bởi dấu cách:")
    for i in range(n):
        hang = list(map(int, input(f"Hàng {i+1}: ").split()))
        mang.append(hang)
    cheo_chinh = [mang[i][i] for i in range(n)]
    cheo_phu = [mang[i][n - 1 - i] for i in range(n)]
    print(f"\nĐường chéo chính: {cheo_chinh}")
    print(f"Đường chéo phụ: {cheo_phu}")
W6A17()
```

```python
def W6A18():
    n, m, k = map(int, input("Nhập n, m, k (cách nhau dấu cách): ").split())
    mang = []
    print(f"Nhập mảng {n}x{m}:")
    for i in range(n):
        hang = list(map(int, input(f"Hàng {i+1}: ").split()))
        mang.append(hang)
    k_index = k - 1
    if 0 <= k_index < n:
        tong_hang_k = sum(mang[k_index])
        print(f"Tổng các phần tử trên HÀNG {k}: {tong_hang_k}")
    if 0 <= k_index < m:
        tong_cot_k = sum(mang[i][k_index] for i in range(n))
        print(f"Tổng các phần tử trên CỘT {k}: {tong_cot_k}")
W6A18()
```

```python
def W6A21():
    chuoi_input = input("Nhập một dòng các chuỗi (cách nhau bởi dấu cách): ")
    danh_sach_chuoi = chuoi_input.split()
    
    dict_vi_tri = {}
    
    for index, chuoi in enumerate(danh_sach_chuoi):
        if chuoi not in dict_vi_tri:
            dict_vi_tri[chuoi] = []
        dict_vi_tri[chuoi].append(index)
        
    dict_ket_qua = {chuoi: tuple(vi_tri) for chuoi, vi_tri in dict_vi_tri.items()}
    
    print(f"Dictionary kết quả: {dict_ket_qua}")
W6A21()
```
