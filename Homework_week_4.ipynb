```python
a , b = map(int, input("Enter two numbers: ").split())
total = 0
for i in range(a, b + 1):
    if i < 2:
        continue
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            break
        else:
            total += i
print("The sum of all prime numbers between", a, "and", b, "is:", total)         
                

```

```python
def largest_prime_divisor(n):
    """Find the largest prime divisor efficiently"""
    if n < 2:
        return None
    
    largest = None
    
    while n % 2 == 0:
        largest = 2
        n //= 2
    
    # Check odd numbers from 3 onwards
    i = 3
    while i * i <= n:
        while n % i == 0:
            largest = i
            n //= i
        i += 2
    
    # If n is still greater than 1, then it's a prime factor
    if n > 1:
        largest = n
    
    return largest

# Main program
n = int(input("Enter a number: "))

if n < 2:
    print(f"{n} has no prime divisors.")
else:
    result = largest_prime_divisor(n)
    
    if result == n:
        print(f"{n} is the largest prime divisor of itself (it's prime).")
    else:
        print(f"The largest prime divisor of {n} is: {result}")

# Show example with factorization
print(f"\nHow it works: Repeatedly divide {n} by its prime factors from smallest to largest.")
print("The last prime factor found is the largest prime divisor.")
```
