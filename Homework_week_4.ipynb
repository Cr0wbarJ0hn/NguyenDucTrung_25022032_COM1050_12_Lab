```python
a , b = map(int, input("Enter two numbers: ").split())
total = 0
for i in range(a, b + 1):
    if i < 2:
        continue
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            break
        else:
            total += i
print("The sum of all prime numbers between", a, "and", b, "is:", total)         
```

```python
n = int(input("Enter a number: "))
def check_is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
for j in range(n+1, 2 , -1):
    if n % j == 0 and check_is_prime(j):
        print("The largest prime factor of", n, "is:", j)
        break
```

```python
n = int(input("Enter a number: "))
iterations = 0
while n > 0:
    reversed_n = int(str(n)[::-1])
    n += reversed_n
    iterations += 1
    if str(n) == str(n)[::-1]:
        print("The palindrome is:", n)
        print("Number of iterations:", iterations)
    break
```

W4A10

```python
def tinh_do_dai_collatz(k):
    count = 1
    while k > 1:
        if k % 2 == 0:
            k //= 2
        else:
            k = 3 * k + 1
        count += 1
    return count
n = int(input("Nhập số nguyên dương n: "))
best_x = 1 
max_len = 1     
for x in range(1, n + 1):
    current_len = tinh_do_dai_collatz(x)
    if current_len > max_len:
        max_len = current_len
        best_x = x
print(best_x, max_len)
```
