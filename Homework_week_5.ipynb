```python
#W5A1
def Max_of_two(a,b) -> int:
    list = []
    list.extend([a,b])
    return max(list)
x , y = map(int, input(). split())
print("The bigger number is:", Max_of_two(x,y))
```

```python
#W5A2
def swap(a,b) -> tuple[int,int]:
    a,b = b,a
    return a, b
x,y = map(int, input(). split())
print(swap(x,y))
```

```python
#W5A3
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
x = int(input())
print(is_Prime(x))    
```

```python
#W5A4
def is_Perfect_number(n):
    divisors = []
    for i in range (1, n ):
        if n % i == 0:
            divisors.append(i)
    total = sum(divisors)
    return total == n    
x = int(input())
print(is_Perfect_number(x))
```

```python
#W5A5
def k_position(arr,k):
    for i in range (0, len(arr)):
        if arr[i] == k:
            return i + 1 
    return -1

numb_list = list(map(int,input().split()))
k = int(input())
print(k_position(numb_list,k))
```

```python
#W5A6
def Factorial(n):
    mult = 1 
    for i in range(1,n+1):
        mult *=i
    return mult
x = int(input())
print(Factorial(x))
    
```

```python
#W5A7
def calculator(a, b, operator):
    if operator == '+':
        return a + b
    elif operator == '-':
        return a - b
    elif operator == '/':
        if b == 0:
            return "Error: Division by zero"
        return a / b
    elif operator == '*':
        return a * b
    else:
        return "Error: Invalid operator"

a, b, c = input().split()
a = float(a)
b = float(b)
result = calculator(a, b, c)
print(result)
```

```python
#W5A8
def hamming_distance(x, y):
    xor_result = x ^ y
    count = 0
    while xor_result:
        count += xor_result & 1  # Kiểm tra bit cuối cùng
        xor_result >>= 1  # Dịch phải 1 bit
    return count
x, y = map(int, input().split())
print(hamming_distance(x, y))
```

```python
#W5A9
def sum_digits(n):
    total = 0
    while n > 0:
        total += n % 10  
        n //= 10         
    return total

# Test
n = int(input())
print(sum_digits(n))
```

```python
#W5A10
def is_isomorphic(a, b):
    if len(a) != len(b):
        return False
    map_a_to_b = {}
    map_b_to_a = {}
    
    for i in range(len(a)):
        char_a = a[i]
        char_b = b[i]
        
        if char_a in map_a_to_b:
            if map_a_to_b[char_a] != char_b:
                return False
        else:
            map_a_to_b[char_a] = char_b
        
        if char_b in map_b_to_a:
            if map_b_to_a[char_b] != char_a:
                return False
        else:
            map_b_to_a[char_b] = char_a
    
    return True

a, b = input().split()
print(is_isomorphic(a, b))
```
